# Electrons dans un cristal - théorème de Bloch

## Cristal

Un cristal parfait est décrit par un arrangement périodique infini d'atomes. Cet arrangement est caractérisé par **une maille élémentaire** et un **réseau de Bravais**:

$$
\vec{R}_i=n_i \vec{a}_1 + m_i \vec{a}_1 + l_i \vec{a}_3
$$

où $n_i$, $m_i$ et $l_i$ sont des entiers relatifs ($n_i,m_i,l_i \in \mathbb{Z}^3$) et où $\vec{a}_1$, $\vec{a}_2$ et $\vec{a}_3$ sont les **vecteurs de base** du réseau.

## Réseau réciproque

Il est utile, par exemple pour décrire le phénomène de diffraction, d'introduire le réseau réciproque composé des vecteurs $\vec{K}$ tels que 

$$ 
e^{i\vec{K}\cdot\vec{R}_i}=1
$$

pour tous les vecteurs $\vec{R}_i$ du réseau, c'est-à-dire tels qu'une onde plane de vecteur d'onde $\vec{K}$, $e^{i\vec{K}\cdot\vec{r}}$ soit périodique avec la périodicité du cristal. 

Nous allons voir que le réseau réciproque joue un rôle fondamental dans la structure électronique des cristaux. 

Il s'agit lui-même d'un réseau de Bravais, généré par trois vecteurs de base $\vec{b}_i$, $i=1,2,3$ tels que 

$$
\vec{b}_i\cdot \vec{a}_j = 2\pi \delta_{ij}.
$$

On peut donner les expressions suivantes des trois vecteurs de base du réseau réciproque:

$$
\vec{b}_1= 2\pi \frac{\vec{a}_2\wedge \vec{a}_3}{\vec{a}_1\cdot(\vec{a}_2\wedge \vec{a}_3)},
$$

$$
\vec{b}_2= 2\pi \frac{\vec{a}_3\wedge \vec{a}_1}{\vec{a}_2\cdot(\vec{a}_2\wedge \vec{a}_3)},
$$

$$
\vec{b}_3= 2\pi \frac{\vec{a}_1\wedge \vec{a}_2}{\vec{a}_2\cdot(\vec{a}_2\wedge \vec{a}_3)}.
$$

(rappel: le volume $V$ de la maille élémentaire est $V=\vec{a}_1\cdot(\vec{a}_2\wedge \vec{a}_3)$)

### Première zone de Brillouin

Une maille élémentaire particulière du réseau réciproque est la première zone de Brillouin définie comme l'ensemble des points plus proches de l'origine, $\vec{0}$ souvent dénoté comme le point $\Gamma$, que de tout autre vecteur $\vec{K}$ du réseau réciproque.

## Théorème de Bloch

### Enoncé

Les états propres $\psi(\vec{r})$ solutions d'un Hamiltonien à un électron 

$$
\hat{H}=-\frac{\hbar^2 }{2m}\vec{\nabla}^2 + U(\vec{r}),
$$

où $U{\vec{r})$ est un potentiel périodique avec la périodicité du réseau de Bravais:

$$
U(\vec{r}+\vec{R})=U(\vec{r}),
$$

pour tout $\vec{R}$ du réseau de Bravais, s'écrivent sous la forme

$$
\psi(\vec{r})=e^{i\vec{k}\cdot\vec{r}}u_\vec{k}(\vec{r}),
$$

avec $u_\vec{k}(\vec{r})$ une fonction périodique avec la périodicité du réseau de Bravais:

$$
u_\vec{k}(\vec{r}+\vec{R})=u_\vec{k}(\vec{r}).
$$

On vérifie en particulier que

$$
\psi(\vec{r}+\vec{R})=e^{i\vec{k}\cdot\vec{R}}\psi(\vec{r}).
$$

### Démonstration

On introduit l'opérateur de translation $\hat{T}_\vec{R}$ défini par

$$
\hat{T}_\vec{R} \psi(\vec{r}) = \psi(\vec{r}+\vec{R}).
$$

On vérifie facilement grâce à la périodicité de $U(\vec{r})$ que 

$$
\hat{T}_\vec{R}\hat{H}=\hat{H}\hat{T}_\vec{R},
$$

c'est-à-dire que l'opérateur translation commute avec le hamiltonien. De plus,

\begin{eqnarray}
\hat{T}_\vec{R}\hat{T}_{\vec{R}^\prime}\psi(\vec{r})=&\psi(\vec{r}+\vec{R}+\vec{R}^\prime)\\
= &\hat{T}_{\vec{R}+\vec{R}^\prime} \psi(\vec{r}),
\end{eqnarray}

soit

$$
\hat{T}_\vec{R}\hat{T}_{\vec{R}^\prime} = \hat{T}_{\vec{R}+\vec{R}^\prime}.
$$

On remarque qu'il s'agit donc d'une représentation du groupe des translations de vecteur $\vec{R}$.

Puisque les opérateurs de translation commutent avec $\hat{H}$ et commutent entre eux, il est possible de choisir des vecteurs propres *communs* à tous ces opérateurs, c'est-à-dire vérifiant:

\begin{eqnarray}
\hat{H} \psi(\vec{r}) = & \epsilon \psi(\vec{r}) \\
\hat{T}_\vec{R}\psi(\vec{r}) = & c(\vec{R}) \psi(\vec{r}).
\end{eqnarray}

Des propriétés des opérateurs de translation, on a

$$
c(\vec{R}+\vec{R}^\prime)=c(\vec{R})\times c(\vec{R}^\prime),
$$

et $c(\vec{R}\in \mathbb{C}$ doit vérifier $\vert c(\vec{R})\vert=1$ puisque la norme de $\psi(\vec{r})$ est conservée par la translation. On pose alors

$$
c(\vec{a}_i)=e^{i2\pi x_i}
$$

et on en déduit

$$ 
c(n_1 \vec{a}_1 + n_2 \vec{a}_2 + n_3 \vec{a}_3)= e^{i2\pi (n_1 x_1+n_2 x_2+ n_3 x_3)}
$$

soit

$$
c(\vec{R})=e^{i2\pi \vec{k}\cdot\vec{R}}
$$

avec

\begin{eqnarray}
\vec{R}=& n_1 \vec{a}_1 + n_2 \vec{a}_2 + n_3 \vec{a}_3 \\
\vec{k}=& x_1 \vec{b}_1 + x_2 \vec{b}_2 + x_3 \vec{b}_3.
\end{eqnarray}

On a alors bien le résultat cherché: on peut choisir les états propres $\psi(\vec{r})$ de $\hat{H}$ tels que

$$
\psi(\vec{r}+\vec{R})=e^{i2\pi \vec{k}\cdot\vec{R}}\psi(\vec{r}).
$$

### Remarques

1. $\vec{k}$ peut toujours être choisi dans la première zone de Brillouin. En effet, si $\vec{K}$ est un vecteur du réseau réciproque et $u(\vec{r})$ une fonction périodique, $e^{i\vec{K}\cdot\vec{r}}u(\vec{r})$ est encore périodique et en écrivant

$$
e^{i(\vec{k}+\vec{K})\cdot\vec{r}}u(\vec{r})= e^{i\vec{k}\cdot\vec{r}}\times e^{i\vec{K}\cdot\vec{r}}u(\vec{r})
$$

on peut toujours ramener $\vec{k}$ dans la première zone de Brillouin.


2. Soit $\psi_\vec{k}(\vec{r})=e^{i\vec{k}\cdot\vec{r}}u_\vec{k}(\vec{r})$ un vecteur propre de $\hat{H}$:

$$
\hat{H}\psi_\vec{k}(\vec{r})=\epsilon_k \psi_\vec{k}(\vec{r}).
$$

On a alors

\begin{eqnarray}
\hat{H}e^{i\vec{k}\cdot\vec{r}}u_\vec{k}(\vec{r})=& -\frac{\hbar^2 }{2m}\vec{\nabla}^2 e^{i\vec{k}\cdot\vec{r}}u_\vec{k}(\vec{r})+ U(\vec{r})e^{i\vec{k}\cdot\vec{r}}u_\vec{k}(\vec{r})\\
=& e^{i\vec{k}\cdot\vec{r}}\left[\frac{\hbar^2 }{2m}(-i\vec{\nabla}+\vec{k})^2+U(\vec{r})\right]u_\vec{k}(\vec{r}).
\end{eqnarray}

$u_\vec{k}(\vec{r})$ est donc solution de 

$$
\hat{H}_\vec{k} u_\vec{k}(\vec{r}) = \epsilon_\vec{k} u_\vec{k}(\vec{r})
$$

avec

$$
\hat{H}_\vec{k}=\frac{\hbar^2 }{2m}(-i\vec{\nabla}+\vec{k})^2+U(\vec{r}).
$$

En général cet hamiltonien a des valeurs propres discrètes car $u_\vec{k}(\vec{r})$ est périodique. On peut alors indexer les solutions par un entier $n$

$$
\hat{H}_\vec{k} u_{n\vec{k}}(\vec{r}) = \epsilon_n(\vec{k}) u_{n\vec{k}}(\vec{r})
$$

ou

$$
\hat{H} \psi_{n\vec{k}}(\vec{r}) = \epsilon_n(\vec{k}) \psi_{n\vec{k}}(\vec{r})
$$

avec 

$$
\psi_{n\vec{k}}(\vec{r})=e^{i\vec{k}\cdot\vec{r}}u_{n\vec{k}}(\vec{r}).
$$


3. Bien que $\vec{k}$ puisse être ramené dans la première zone de Brillouin, il est naturel d'étendre $\vec{k}$ par périodicité sur tout l'espace par:

$$
\epsilon_n(\vec{k}+\vec{K})=\epsilon_n(\vec{k}).
$$

Comme on s'attend à ce que $\epsilon_n(\vec{k})$ soit une fonction continue de $\vec{k}$, alors $\epsilon_n(\vec{k})$ possède un maximum et un mininum: on parle alors de **bande d'énergie**.


## Illustration à une dimension avec le polyacétylène

### Présentation du modèle

On représente le polyacétylène comme un polymère 1D parfait, suite périodique de groupements CH$_2$ avec une alternance simple liaison, double liaison. La maille élémentaire est alors constituée de deux groupements. On distinguera généralement ces deux groupements A et B. Cela permet par exemple la généralisation du problème à un système avec une alternance de type BN ou alors avec des substitutions périodiques des groupements CH2, par exemple CH$_2$-CCl$_2$... Enfin, nous n'allons considérer ici que les électrons $p_z$ de ces groupements avec donc un électron par groupement, soit deux électrons par maille. On notera $d$ le paramètre de maille.

On note $\vert A,n\rangle$ l'orbitale $p_z$ du groupement A de la maille $n$ et de même $\vert B,n\rangle$ pour le groupement B. On considère alors un modèle "à la Hückel", appelé dans ce contexte modèle de *liaison forte*. On pose alors

$$
\langle A,n \vert \hat{H} \vert A,n \rangle = \alpha_1,
$$

$$
\langle B,n \vert \hat{H} \vert B,n \rangle = \alpha_2,
$$

$$
\langle A,n \vert \hat{H} \vert B,n \rangle = \beta_1,
$$

$$
\langle B,n \vert \hat{H} \vert A,n+1 \rangle = \beta_2,
$$

où $\alpha_1 < 0$, $\alpha_2 < 0 $ sont les énergies de site des orbitales $p_z$, et $\beta_1 < 0$ et $\beta_2 < 0$ sont les couplages entre les orbitales $p_z$ de deux groupements voisins. Au delà des premiers voisins, les couplages sont nuls:

$$
\langle A,n \vert \hat{H} \vert B,m \rangle = 0\text{, si $m>n$},
$$

$$
\langle B,n \vert \hat{H} \vert A,m \rangle = 0\text{, si $m>n+1$},
$$

et

$$
\langle A,n \vert \hat{H} \vert A,m \rangle = \langle B,n \vert \hat{H} \vert B,m \rangle = 0\text{, si $m\neq n$}.
$$


### Zone de Brillouin

Le réseau réciproque est alors un réseau 1D de maille $\frac{2\pi}{d}$. La première zone de Brillouin est alors le segment  $-\frac{\pi}{d}< k < \frac{\pi}{d}$.

### Orbitales monoélectroniques

Une orbitale électronique s'écrit de manière générale comme une combinaison linéaire des orbitales $p_z$:

$$
\vert \varPsi \rangle = \frac{1}{\sqrt{N}}\sum_n ( c_n \vert A,n\rangle + d_n \vert B,n \rangle),
$$

où $N$ est le nombre de monomères du polymère et avec

$$
\frac{1}{N}\sum_n \vert c_n \vert^2 + \vert d_n \vert^2 =1.
$$

D'après le théorème de Block, il est alors possible d'écrire les états propres du hamiltonien $\hat{H}$ comme

$$
\vert \varPsi_k \rangle =  \frac{1}{\sqrt{N}} \sum_n e^{ikdn}  c(k) \vert A,n\rangle + e^{ik(dn+d/2)}  d(k) \vert B,n \rangle,
$$

avec $-\frac{\pi}{d}< k < \frac{\pi}{d}$. Ceci correspond à une orbitale périodique 

$$
\vert u_k \rangle = \frac{1}{\sqrt{N}} \sum_n  c(k) \vert A,n\rangle + d(k) \vert B,n \rangle
$$ 

(en "représentation $n$" on écrirait $u_k(A,n)=c(k)$, $u_k(B,n)=d(k)$), avec $\vert c(k)\vert^2 + \vert d(k)\vert^2=1$.

Remarquez que par commodité cela revient à placer le groupement $B$ au milieu de la maille (à une distance $d/2$ de A). Ceci n'a aucun rôle à la fin car ce n'est qu'une autre définition du coefficient $d(k)$.


### Résolution de l'équation de Schrödinger indépendante du temps

L'équation de Schrödinger indépendante du temps dont $\vert \varPsi_k \rangle$ est solution s'écrit

$$
\hat{H}\vert \varPsi_k \rangle = \epsilon(k) \vert \varPsi_k \rangle.
$$

D'après le modèle décrit ci-dessus, on a 

$$
\hat{H}\vert A,n \rangle = \alpha_1 \vert A,n \rangle + \beta_1 \vert B,n \rangle +\beta_2 \vert B,n-1\rangle
$$

et

$$
\hat{H}\vert B,n \rangle = \alpha_2 \vert B,n \rangle + \beta_1 \vert A,n \rangle +\beta_2 \vert A,n+1\rangle.
$$

Soit à partir de l'expression de $\vert \varPsi_k \rangle$ prescrite par le théorème de Bloch:

$$
\hat{H}\vert \varPsi_k \rangle = \frac{1}{\sqrt{N}}\sum_n e^{ikdn}  c(k)\left( \alpha_1 \vert A,n \rangle + \beta_1 \vert B,n \rangle +\beta_2 \vert B,n-1\rangle \right)+ e^{ik(dn+d/2)}  d(k) \left(
\alpha_2 \vert B,n \rangle + \beta_1 \vert A,n \rangle +\beta_2 \vert A,n+1\rangle \right)
$$

Après quelques manipulations, on obtient

$$
\hat{H}\vert \varPsi_k \rangle = \frac{1}{\sqrt{N}}\sum_n e^{ikdn} \left( \alpha_1 c(k) + e^{ikd/2} \beta_1 d(k) + e^{-ikd/2} \beta_2 d(k)\right) \vert A,n \rangle + e^{ik(dn+d/2)} \left( e^{-ikd/2} \beta_1 c(k) + e^{ikd/2} \beta_2 c(k) + \alpha_2 d(k) \right) \vert B,n \rangle.
$$

$\vert u_k \rangle$ est alors solution de l'équation aux valeurs propres paramétrée par $k$:

$$
\hat{H}_k \vert u_k \rangle = \frac{1}{\sqrt{N}}\sum_n  \left( \alpha_1 c(k) + e^{ikd/2} \beta_1 d(k) + e^{-ikd/2} \beta_2 d(k)\right) \vert A,n \rangle +  \left( e^{-ikd/2} \beta_1 c(k) + e^{ikd/2} \beta_2 c(k) + \alpha_2 d(k) \right) \vert B,n \rangle = \epsilon(k) \vert u_k \rangle.
$$

En représentant $\vert u_k \rangle$ par le vecteur colonne $\begin{pmatrix} c(k) \\ d(k) \end{pmatrix}$, celui-ci vérifie l'équation aux valeurs propres:

$$
H_k \begin{pmatrix} c(k) \\ d(k) \end{pmatrix} = 
\begin{pmatrix}
\alpha_1 & e^{ikd/2} \beta_1  + e^{-ikd/2} \beta_2 \\
e^{-ikd/2} \beta_1 + e^{ikd/2} \beta_2 & \alpha_2
\end{pmatrix}
\begin{pmatrix} c(k) \\ d(k) \end{pmatrix} = \epsilon(k) \begin{pmatrix} c(k) \\ d(k) \end{pmatrix}
$$

### Valeurs propres et structure de bandes

Il est possible de résoudre analytiquement les valeurs propres de cette matrice $2\times 2$ et on obtient

$$
\epsilon_+(k)=\frac{\alpha_1 + \alpha_2}{2} + \sqrt{\frac{(\alpha_1-\alpha_2)^2}{4}+(\beta_1^2+\beta_2^2+2 \beta_1\beta_2 \cos kd)}
$$

et

$$
\epsilon_-(k)=\frac{\alpha_1 + \alpha_2}{2} - \sqrt{\frac{(\alpha_1-\alpha_2)^2}{4}+(\beta_1^2+\beta_2^2+2 \beta_1\beta_2 \cos kd)}
$$


### Illustration

```python
# modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# definition de H_k comme matrice 2x2

def hamitonian(alpha1,alpha2,beta1,beta2,d,k):
    ham=np.zeros((2,2),dtype=complex)
    ham[0,0]=alpha1
    ham[1,1]=alpha2
    ham[0,1]=beta1*np.exp(1j*k*d/2)+beta2*np.exp(-1j*k*d/2)
    ham[1,0]=np.conj(ham[0,1])
    return ham
```

```python
# paramètres

d=2
alpha1,alpha2=-1.0,-1.0
beta1,beta2=-1.5,-0.5
```

```python
# Nbre de points en k 

N=20
```

```python
# Résolution numérique par numpy

eps=np.zeros((N+1,2))
kpt=np.linspace(-np.pi/d,np.pi/d,N+1)

for ik,k in enumerate(kpt):
    H=hamitonian(alpha1,alpha2,beta1,beta2,d,k)
    e,v=np.linalg.eigh(H)
    eps[ik,:]=e
```

```python
# Solution analytique

def eplus(k):
    return (alpha1+alpha2)/2 + np.sqrt((alpha1-alpha2)**2/4+(beta1**2+beta2**2+2*beta1*beta2*np.cos(k*d)))

def emoins(k):
    return (alpha1+alpha2)/2 - np.sqrt((alpha1-alpha2)**2/4+(beta1**2+beta2**2+2*beta1*beta2*np.cos(k*d)))
```

```python
# Figure

plt.plot(kpt,eps[:,0],'ro',label=r'$\epsilon_-(k)$ numérique')
plt.plot(kpt,eps[:,1],'bo',label=r'$\epsilon_+(k)$ numérique')

plt.plot(kpt,eplus(kpt),'b',label=r'$\epsilon_+(k)$ analytique')
plt.plot(kpt,emoins(kpt),'r',label=r'$\epsilon_-(k)$ analytique')

plt.xlabel(r'$k$')
plt.ylabel(r'$\epsilon(k)$')

plt.legend()
plt.show()
```

### Nombres de vecteurs $k$ et nombre d'électrons: conditions aux bords du cristal

Maintenant que l'on a déterminé les énergies propres du système, il nous faut connaître le remplissage des niveaux d'énergie. Si on se réfère à la statistique de Fermi-Dirac et si on se place à température nulle, tous les niveaux 
d'énergie inférieure à l'énergie de Fermi sont occupés par $2s+1$ électrons. Le niveau de Fermi est alors déterminé par le nombre d'électrons dans le système. Comme pour le gaz parfait d'électrons, il nous faut alors discuter des conditions aux bords.

En reprenant ce que l'on a fait pour le gaz d'électrons, on considère un cristal de $N_1$ cellules en $\vec{a}_1$, $N_2$ cellules en $\vec{a}_2$ et $N_3$ cellules en $\vec{a}_3$. Le système comporte alors $N=N_1\times N_2 \times N_3$ cellules et a un volume $V_{tot}=NV$ où $V$ est le volume de la maille élémentaire.
